```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_excel("TD_PE.xlsx")
df
```

```python
df = df.rename(columns={'PE Ratio': 'PE', 'TTM Net EPS': 'EPS', 'Stock Price': 'stock_price'})
df.head()
```

```python
alt.Chart(df).mark_line().encode(
    x=alt.X('Date:T', title='Date', axis=alt.Axis(format='%Y-%m-%d', labelAngle=-45)),
    y=alt.Y('PE:Q', title='PE Ratio'),
    color = alt.Color('Bank:N')
).properties(width=600, height=200, title="TD PE Ratio vs. Other Bank's PE")
```

```python
td_data = df[df['Bank'] == 'TD']
other_data = df[df['Bank'] != 'TD']

average_pe_other = other_data.groupby('Date')['PE'].mean().reset_index()

line_td = alt.Chart(td_data).mark_line(color='blue').encode(
    x=alt.X('Date:T', title='Date', axis=alt.Axis(format='%Y-%m-%d', labelAngle=-45)),
    y=alt.Y('PE:Q', title='PE Ratio'),
    tooltip=['Date:N', 'PE:Q'],
    color=alt.Color('Bank:N')
).properties(
    title='TD PE Ratio vs. Industry PE Average',
    width=600,
    height=200
)

line_average_other = alt.Chart(average_pe_other).mark_line(color='orange').encode(
    x=alt.X('Date:T', title='Date', axis=alt.Axis(format='%Y-%m-%d', labelAngle=-45)),
    y=alt.Y('PE:Q', title='PE Ratio'),
    tooltip=['Date:N', 'PE:Q'],
)

# Combine the charts
chart = line_td + line_average_other

# Show the chart
chart
```
